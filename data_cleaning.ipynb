```python
import pandas as pd
import numpy as np

# Load dataset

df = pd.read_csv("marketing_campaign.csv", sep="\t")
print(df.head())
```

```python
# Basic Information
df.info()
df.describe()
```

```python
# Check Missing Values
df.isnull().sum()
```

```python
# Fill missing numerical values with median
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
# Remove Duplicate
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Rename Columns
df.columns = (df.columns
      .str.lower()
      .str.strip()
      .str.replace(" ", "_")
)
```

```python
df.info()
```

```python
# Standardize Text Columns

text_columns = ["education", "marital_status"]

for col in text_columns:
    df[col] = df[col].str.strip().str.title()

print(df.head(10))
```

```python
df["dt_customer"] = pd.to_datetime(
    df["dt_customer"],
    format="%d-%m-%Y"
)

df["dt_customer"] = df["dt_customer"].dt.strftime("%d-%m-%Y")
print(df.head(3))
```
